### Getting Started

Before you start cleaning a data set, it’s a good idea to just get a general feel for the data. After that, you can put together a plan to clean the data.

I like to start by asking the following questions:

- What are the features?

- What are the expected types (int, float, string, boolean)?

- Is there obvious missing data (values that Pandas can detect)?

- Is there other types of missing data that’s not so obvious (can’t easily detect with Pandas)?

To show you what I mean, let’s start working through the example.

### Unexpected Missing Values

```python
# Looking at the OWN_OCCUPIED column
print(df['OWN_OCCUPIED'])
print(df['OWN_OCCUPIED'].isnull())
```

This example is a little more complicated so we’ll need to think through a strategy for detecting these types of missing values. There’s a number of different approaches, but here’s the way that I’m going to work through this one.

1. Loop through the OWN_OCCUPIED column

2. Try and turn the entry into an integer

3. If the entry can be changed into an integer, enter a missing value

4. If the number can’t be an integer, we know it’s a string, so keep going

Let’s take a look at the code and then we’ll go through it in detail.

```python
# Detecting numbers in OWN_OCCUPIED column
cnt=0
for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        df.loc[cnt, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
```

```python
# coerce NUM_BATH to numeric
df['NUM_BATH'] = pd.to_numeric(df['NUM_BATH'], errors='coerce')
```

```python
df.head(10)
```

You’ll notice that I used `try` and except `ValueError`. This is called <u>exception handling</u>, and we use this to handle errors.

If we were to try and change an entry into an integer and it couldn’t be changed, then a `ValueError` would be returned, and the code would stop. To deal with this, we use exception handling to recognize these errors, and keep going.
Another important bit of the code is the `.loc` method. This is the preferred Pandas method for modifying entries in place.

### Replacing Missing Values

```python
# Replace missing values with a number
df['PID'].fillna(100005000, inplace=True)
df['ST_NUM'].fillna(125, inplace=True)
```

```python
# Location based replacement
df.loc[6,'ST_NUM'] = 151
```

```python
# Replace using median 
median = df['NUM_BEDROOMS'].median()
df['NUM_BEDROOMS'].fillna(median, inplace=True)
```

```python
# Replace using mode 
df['OWN_OCCUPIED'].fillna(df['OWN_OCCUPIED'].mode()[0], inplace=True)
```

```python
df.head(10)
```

### Imputing numeric values

```python
# let's visualize NUM_BEDROOMS and SQ_FT
sns.regplot(x=df['NUM_BEDROOMS'], y=df['SQ_FT'])
```

```python
# correlation between NUM_BEDROOMS adn SQ_FT
df['NUM_BEDROOMS'].corr(df['SQ_FT'])
```

```python
# let's visualize NUM_BEDROOMS and NUM_BATH
sns.regplot(x=df['NUM_BEDROOMS'], y=df['NUM_BATH'])
```

```python
# correlation between NUM_BEDROOMS adn NUM_BATH
df['NUM_BEDROOMS'].corr(df['NUM_BATH'])
```

```python
# impute NUM_BATH and SQ_FT with a simple regression with x = NUM_BEDROOMS
# Source: https://www.kaggle.com/shashankasubrahmanya/missing-data-imputation-using-regression
missing_columns = ['NUM_BATH', 'SQ_FT']
```

```python
# create _imp columns
for feature in missing_columns:
    df[feature + '_imp'] = df[feature]
```

```python
# replace NAN with 0 on _imp columns
df['NUM_BATH_imp'].fillna(0, inplace=True)
df['SQ_FT_imp'].fillna(0, inplace=True)
```

```python
df.head(10)
```

```python
# Create a Linear Regression model to estimate the missing data
model = linear_model.LinearRegression()

X = df.iloc[:, 4].values.reshape(-1, 1)  # values converts it into a numpy array

for feature in missing_columns:
    model.fit(X, y = df[feature + '_imp'])
    df.loc[df[feature].isnull(), feature + '_imp'] = model.predict(X)[df[feature].isnull()]
```

```python
df.head(10)
```

```python
# replace NUM_BATH equal 1 if less than 1
df.loc[df.NUM_BATH_imp < 1, 'NUM_BATH_imp'] = 1
```

```python
df.head(10)
```

```python
# replace NUM_BATH and SQ_FT with _imp columns values
df_imp = df.copy()
df_imp['NUM_BATH'] = df_imp['NUM_BATH_imp']
df_imp['SQ_FT'] = df_imp['SQ_FT_imp']
```

```python
df_imp.drop(['NUM_BATH_imp', 'SQ_FT_imp'], axis=1, inplace=True)
```
